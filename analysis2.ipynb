```python

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python

df = pd.read_csv('data.csv')
# df.head()
df.drop('Unnamed: 0', axis=1, inplace=True)
df['Num_Ratings'] = df['Num_Ratings'].str.replace(',', '').astype(float)
print(df.head())
```

```python

# Basic statistics
basic_stats = df.describe()

# Average Ratings Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Avg_Rating'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Распределение средних оценок')
plt.xlabel('Средний рейтинг')
plt.ylabel('Частота')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Top 10 Authors by Number of Books
top_authors = df['Author'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_authors.plot(kind='barh', color='steelblue')
plt.title('Топ-10 авторов по количеству книг')
plt.xlabel('Количество книг')
plt.ylabel('Автор')
plt.show()

# Number of Books in Each Genre
all_genres = pd.Series([genre for sublist in df['Genres'].dropna().apply(eval) for genre in sublist])
top_genres = all_genres.value_counts().head(10)
plt.figure(figsize=(10, 6))
top_genres.plot(kind='bar', color='teal')
plt.title('Топ-10 жанров по количеству книг')
plt.xlabel('Жанр')
plt.ylabel('Количество книг')
plt.show()
```

```python

# Display the first few rows of the DataFrame
print(df.head())

# Visualization of the number of books per author in the top 10 most prolific authors
top_authors = df['Author'].value_counts().head(10)
plt.figure(figsize=(12, 8))
top_authors.plot(kind='bar', color='purple')
plt.title('Топ-10 самых плодовитых авторов')
plt.xlabel('Автор')
plt.ylabel('Количество книг')
plt.xticks(rotation=45)
plt.show()

# # Boxplot of Average Ratings
# plt.figure(figsize=(10, 6))
# plt.boxplot(df['Avg_Rating'].dropna(), vert=False)
# plt.title('Блок-схема средних оценок')
# plt.xlabel('Средний рейтинг')
# plt.show()

# Display a table of basic statistics for numerical columns
print(df.describe())
```
